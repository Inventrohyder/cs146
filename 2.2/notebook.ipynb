```python
# Setup
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Generate X points
x = np.linspace(-5, 5, 200)  # Array with 200 linearly spaced points in [-5,5]
```

```python
# Gamma distributions
alpha_1 = 1
alpha_2 = 2

beta = 1

plt.figure(figsize=(8, 4))  # Create a new figure with a particular size
plt.plot(x, stats.gamma.pdf(x, a=alpha_1, scale=1/beta), label=f'alpha = {alpha_1}')
plt.plot(x, stats.gamma.pdf(x, a=alpha_2, scale=1/beta), label=f'alpha = {alpha_2}')
plt.legend()  # Show the legend in the top right corner
plt.title('Gamma distributions')
plt.show()
```

```python
# Normal distributions
sigma_1 = 2.5
sigma_2 = 5

mu = 0

plt.figure(figsize=(8, 4))  # Create a new figure with a particular size
plt.plot(x, stats.norm.pdf(x, loc=mu, scale=sigma_1), label=f'scale = {sigma_1}')
plt.plot(x, stats.norm.pdf(x, loc=mu, scale=sigma_2), label=f'scale = {sigma_2}')
plt.legend()  # Show the legend in the top right corner
plt.title('Normal distributions')
plt.show()
```
